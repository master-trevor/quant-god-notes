## Step 1 ‚Äî Binomial pmf  
Write the pmf for $X \sim \text{Binomial}(n,p)$.

- $P(X=k) \;=\; \binom{n}{k} \, p^k \, q^{\,n-k}, \quad q = 1-p$

---

## Step 2 ‚Äî Stirling‚Äôs Approximation  
Apply Stirling‚Äôs formula to $n!$, $k!$, $(n-k)!$ inside $\binom{n}{k}$.

- $m! \;\approx\; \sqrt{2\pi m}\,\Big(\tfrac{m}{e}\Big)^m$

So:

- $\displaystyle \binom{n}{k} \;\approx\; \frac{1}{\sqrt{2\pi}} \sqrt{\frac{n}{k(n-k)}} \; \frac{n^n}{k^k (n-k)^{\,n-k}}$

Now substitute this into the pmf **with $p^k q^{\,n-k}$ still present**:

- $\displaystyle P(X=k) \;\approx\; \frac{1}{\sqrt{2\pi}} \sqrt{\frac{n}{k(n-k)}} \;\frac{n^n}{k^k (n-k)^{\,n-k}} \; p^k q^{\,n-k}$

---

## Step 3 ‚Äî Reorganize terms  
Combine $n^n$ with $p^k$ and $q^{\,n-k}$ to form ratios:

- $\displaystyle P(X=k) \;\approx\; \frac{1}{\sqrt{2\pi}} \sqrt{\frac{n}{k(n-k)}} \; \Big(\frac{np}{k}\Big)^k \; \Big(\frac{nq}{n-k}\Big)^{\,n-k}$

---

## Step 4 ‚Äî Log Transformation  
Take logs to turn products into sums.

- $\ln P \;\approx\; -\tfrac12\ln(2\pi) \;+\; \tfrac12(\ln n - \ln k - \ln(n-k)) \;+\; k \ln\!\Big(\tfrac{np}{k}\Big) \;+\; (n-k)\ln\!\Big(\tfrac{nq}{n-k}\Big)$

---

## Step 5 ‚Äî Centering  
Introduce deviation $\delta$:

- $k = np + \delta, \quad n-k = nq - \delta$

---

## Step 6 ‚Äî First Ratio Expansion  
Expand the first ratio term:

- $\ln\!\Big(\tfrac{np}{k}\Big) = \ln\!\Big(\tfrac{np}{np+\delta}\Big) = -\ln\!\Big(1+\tfrac{\delta}{np}\Big)$

Use Taylor: $\ln(1+x) \approx x - \tfrac{x^2}{2}$ for small $x$.

Then multiply by $k=np+\delta$.

- Fill in: $k \ln\!\Big(\tfrac{np}{k}\Big) \;\approx\; \_$

---

## Step 7 ‚Äî Second Ratio Expansion  
Expand the second ratio term:

- $\ln\!\Big(\tfrac{nq}{n-k}\Big) = \ln\!\Big(\tfrac{nq}{nq-\delta}\Big) = -\ln\!\Big(1-\tfrac{\delta}{nq}\Big)$

Use expansion: $-\ln(1-x) \approx x + \tfrac{x^2}{2}$.

Then multiply by $(n-k)=nq-\delta$.

- Fill in: $(n-k)\ln\!\Big(\tfrac{nq}{n-k}\Big) \;\approx\; \_$

---

## Step 8 ‚Äî Combine  
Add Step 6 and Step 7 results. Show linear terms cancel; keep quadratic order.

- Result: $\;\approx\; -\tfrac{\delta^2}{2np} - \tfrac{\delta^2}{2nq} \;=\; -\tfrac{\delta^2}{2npq}$

---

## Step 9 ‚Äî Square-root Prefactor  
Simplify:

- $\tfrac12[\ln n - \ln k - \ln(n-k)] \;\approx\; -\tfrac12 \ln(npq)$

---

## Step 10 ‚Äî Collect Everything  
Write the final expression for $\ln P$:

- $\ln P \;\approx\; -\tfrac12\ln(2\pi npq) - \tfrac{(k-np)^2}{2npq}$

---

## Step 11 ‚Äî Exponentiate  
Undo the log to recover $P(X=k)$:

- $\displaystyle P(X=k) \;\approx\; \frac{1}{\sqrt{2\pi \, npq}} \; \exp\!\Big(-\tfrac{(k-np)^2}{2\,npq}\Big)$
