## Step 4 ‚Äî Centering  
Introduce deviation $\delta$ by setting $k = np + \delta$, so $n-k = nq - \delta$.

- $k = \_$  
- $n-k = \_$

## Step 5 ‚Äî First Ratio Expansion  
Rewrite and expand:

- $\ln\!\left(\tfrac{np}{k}\right) = \ln\!\left(\tfrac{np}{np+\delta}\right) = -\ln\!\left(1 + \tfrac{\delta}{np}\right)$  

Use the Taylor expansion $\ln(1+x) \approx x - \tfrac{x^2}{2}$ (for small $x$).  
Then multiply by $k=np+\delta$.

- Result: $k\ln\!\left(\tfrac{np}{k}\right) \;\approx\; \_$

---

## Step 6 ‚Äî Second Ratio Expansion  
Similarly:

- $\ln\!\left(\tfrac{nq}{n-k}\right) = \ln\!\left(\tfrac{nq}{nq-\delta}\right) = -\ln\!\left(1 - \tfrac{\delta}{nq}\right)$  

Use expansion $-\ln(1-x) \approx x + \tfrac{x^2}{2}$.  
Then multiply by $(n-k)=nq-\delta$.

- Result: $(n-k)\ln\!\left(\tfrac{nq}{n-k}\right) \;\approx\; \_$

---

## Step 7 ‚Äî Combine  
Add the two expansions. Show linear terms cancel and quadratics combine.

- Combined: $\;\approx -\tfrac{\delta^2}{2np} - \tfrac{\delta^2}{2nq}$  
- Simplify: $\;\approx -\tfrac{\delta^2}{2npq}$

---

## Step 8 ‚Äî Square-root Prefactor  
Simplify the leftover term:

- $\tfrac12[\ln n - \ln k - \ln(n-k)] \;\approx\; -\tfrac12\ln(npq)$

---

## Step 9 ‚Äî Collect Everything  
Assemble the full expression for $\ln P$.

- $\ln P \;\approx\; -\tfrac12\ln(2\pi npq) - \tfrac{(k-np)^2}{2npq}$

---

## Step 10 ‚Äî Exponentiate  
Undo the log to recover $P(X=k)$.

- $P(X=k) \;\approx\; \dfrac{1}{\sqrt{2\pi npq}} \exp\!\Big(-\tfrac{(k-np)^2}{2npq}\Big)$
