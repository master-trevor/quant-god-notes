# üìù Gaussian-from-Binomial Derivation Worksheet (with explicit Taylor steps)

---

## Step 1 ‚Äî Binomial pmf
Write the pmf for $X \sim \mathrm{Binomial}(n,p)$ with $q=1-p$.

- $P(X=k) \;=\; \binom{n}{k}\,p^k\,q^{\,n-k}$

---

## Step 2 ‚Äî Stirling‚Äôs approximation (apply **only** to factorials)
State Stirling and substitute inside $\binom{n}{k}$ (keep $p^k q^{\,n-k}$ untouched).

- $m! \;\approx\; \sqrt{2\pi m}\,\Big(\tfrac{m}{e}\Big)^m$

- $\displaystyle \binom{n}{k}
\;\approx\;
\frac{1}{\sqrt{2\pi}}\,
\sqrt{\frac{n}{k(n-k)}}\;
\frac{n^n}{k^k (n-k)^{\,n-k}}$

So the pmf becomes

- $\displaystyle P(X=k)
\;\approx\;
\frac{1}{\sqrt{2\pi}}\,
\sqrt{\frac{n}{k(n-k)}}\;
\frac{n^n}{k^k (n-k)^{\,n-k}}\;
p^k\,q^{\,n-k}$

---

## Step 3 ‚Äî Reorganize into ratio form
Group the powers to get clean ratios:

- $\displaystyle P(X=k)
\;\approx\;
\frac{1}{\sqrt{2\pi}}\,
\sqrt{\frac{n}{k(n-k)}}\;
\Big(\frac{np}{k}\Big)^{\!k}\;
\Big(\frac{nq}{\,n-k\,}\Big)^{\!n-k}$

---

## Step 4 ‚Äî Take logs
Turn products into sums:

- $\displaystyle \ln P
\;\approx\;
-\tfrac12\ln(2\pi)
\;+\;\tfrac12\big[\ln n-\ln k-\ln(n-k)\big]
\;+\;k\ln\!\Big(\tfrac{np}{k}\Big)
\;+\;(n-k)\ln\!\Big(\tfrac{nq}{n-k}\Big)$

---

## Step 5 ‚Äî Center around the mean
Introduce deviation $\delta$:

- $k = np + \delta,\qquad n-k = nq - \delta$

---

## Step 6 ‚Äî **Taylor expansion**: first ratio term
Start:
- $\displaystyle \ln\!\Big(\tfrac{np}{k}\Big)
= \ln\!\Big(\tfrac{np}{np+\delta}\Big)
= -\,\ln\!\Big(1+\tfrac{\delta}{np}\Big)$

**Use Taylor near $0$**:
- $\displaystyle \ln(1+x)\;\approx\;x-\tfrac{x^2}{2} \quad (\text{ignore }O(x^3))$

Therefore:
- $\displaystyle \ln\!\Big(\tfrac{np}{k}\Big)
\;\approx\; -\frac{\delta}{np} + \frac{\delta^2}{2n^2p^2}$

Multiply by $k=np+\delta$:
- **Fill:** $\;k\ln\!\Big(\tfrac{np}{k}\Big)\;\approx\;$ ________

*(Target you should reach: $-\delta - \tfrac{\delta^2}{2np}$ up to $O(\delta^2)$.)*

---

## Step 7 ‚Äî **Taylor expansion**: second ratio term
Start:
- $\displaystyle \ln\!\Big(\tfrac{nq}{n-k}\Big)
= \ln\!\Big(\tfrac{nq}{nq-\delta}\Big)
= -\,\ln\!\Big(1-\tfrac{\delta}{nq}\Big)$

**Use Taylor near $0$**:
- $\displaystyle -\ln(1-x)\;\approx\;x+\tfrac{x^2}{2}\quad (\text{ignore }O(x^3))$

Therefore:
- $\displaystyle \ln\!\Big(\tfrac{nq}{n-k}\Big)
\;\approx\; \frac{\delta}{nq} + \frac{\delta^2}{2n^2q^2}$

Multiply by $(n-k)=nq-\delta$:
- **Fill:** $\;(n-k)\ln\!\Big(\tfrac{nq}{n-k}\Big)\;\approx\;$ ________

*(Target you should reach: $\;\delta - \tfrac{\delta^2}{2nq}$ up to $O(\delta^2)$.)*

---

## Step 8 ‚Äî Combine the Taylor pieces (show the cancellation)
Add your Step 6 and Step 7 results:

- Linear in $\delta$: ________ (should cancel to $0$)  
- Quadratic in $\delta$:
  $
  -\frac{\delta^2}{2np} - \frac{\delta^2}{2nq}
  \;=\; -\frac{\delta^2}{2n}\Big(\frac{1}{p}+\frac{1}{q}\Big)
  \;=\; -\frac{\delta^2}{2npq}.
  $

---

## Step 9 ‚Äî Handle the square-root prefactor term
Approximate near $k\approx np$ and $n-k\approx nq$:

- **Fill:** $\;\tfrac12\big[\ln n-\ln k-\ln(n-k)\big]\;\approx\;$ ________

*(Target: $-\tfrac12\ln(npq)$.)*

---

## Step 10 ‚Äî Collect $\ln P$
Assemble everything (Steps 8‚Äì9 plus $-\tfrac12\ln(2\pi)$):

- **Fill:** $\;\ln P\;\approx\;$ ________

*(Target: $-\tfrac12\ln(2\pi npq)\;-\;\dfrac{(k-np)^2}{2npq}$.)*

---

## Step 11 ‚Äî Exponentiate back to $P(X=k)$
Undo the log:

- **Fill:** $\;P(X=k)\;\approx\;$ ________

*(Target:)*
$
\boxed{\,P(X=k)\;\approx\;\frac{1}{\sqrt{2\pi\,npq}}\,
e^{-\frac{(k-np)^2}{2\,npq}}\,}
$

---

### Taylor references (for convenience)
- $\ln(1+x)\approx x-\dfrac{x^2}{2}$, valid for small $|x|$.  
- $-\ln(1-x)\approx x+\dfrac{x^2}{2}$, valid for small $|x|$.
